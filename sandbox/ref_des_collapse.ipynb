```python
stats_data = '/Users/knuth/Documents/ooi/repos/github/ooi_stats/'+ array + '.csv'
deployment_data = pd.read_csv(stats_data)
data = pd.read_csv(stats_data)
```

```python
'''
first, let's change all the deployment values to 0
'''
deployment_data['value'] = 0

'''
next, let's drop all instances where data['value'] == 0 so that the 'data' 
data frame only represents instances where there is data in the system.
'''
data = deployment_data[data.value != 0]

```

```python
'''
notice that value changed to 0
'''
deployment_data
```

```python
'''
and 'data' now only contains instances where data was returned
'''

print(len(data))
print(len(deployment_data))
```

```python
'''
next, let's drop method, stream and value columns from both data frames, then collapse by dropping duplicates
'''
data = data[['refdes', 'deployment','date']]
data = data.drop_duplicates()
deployment_data = deployment_data[['refdes', 'deployment','date']]
deployment_data = deployment_data.drop_duplicates()
```

```python
'''
next, we merge the data frames, using database style merging. note that only 'both and right_only'
merge instances exist, which is expected
'''
deployed_but_no_data = data.merge(deployment_data,indicator=True, how='outer')
deployed_but_no_data = deployed_but_no_data[deployed_but_no_data['_merge'] == 'right_only']
```

```python
deployed_but_no_data
```

```python
'''
next, add back the value columns, delete the '_merge' column and stack the data frames back together
to get availability by instrument and day
'''
del deployed_but_no_data['_merge']
deployed_but_no_data['value'] = 0
data['value'] = 1
```

```python
output = pd.concat([data, deployed_but_no_data])
```

```python
output.to_csv(array+'_refdes.csv', index=False)
```
